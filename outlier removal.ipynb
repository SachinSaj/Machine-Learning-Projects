```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
value = np.random.normal(20000,10000,30000)
```

```python
plt.hist(value,bins=25)
```

```python
val = 2000000
value = np.append(value,val)
plt.hist(value,bins=25)
```

```python
def remove(data):
    median = np.median(data)
    sd = np.std(data)
    new_values = [x for x in data if (median - 1.5*sd < x < median + 1.5*sd)]
    return new_values

nvalue = remove(value)
plt.hist(nvalue,bins=25)
plt.show()
```
